```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

train = pd.read_csv('data/train.csv')
test  = pd.read_csv('data/test.csv')

print ("Trainingsdaten:", train.shape)
print ("Testdaten:", test.shape)
```

```python
train.isnull().sum()
```

```python

train = train.drop(columns=['Cabin', 'Name', 'Ticket', 'PassengerId'])
test  = test.drop(columns=['Cabin', 'Name', 'Ticket'])


train['Age'] = train['Age'].fillna(train['Age'].median())
test['Age']  = test['Age'].fillna(test['Age'].median())


train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])


test['Fare'] = test['Fare'].fillna(test['Fare'].median())

print("Fehlende Werte Train:", train.isnull().sum().sum())
print("Fehlende Werte Test:", test.isnull().sum().sum())
```

```python

train['Age'] = train['Age'].fillna(train['Age'].median())
test['Age']  = test['Age'].fillna(test['Age'].median())


train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])


test['Fare'] = test['Fare'].fillna(test['Fare'].median())

print("Fehlende Werte Train:", train.isnull().sum().sum())
print("Fehlende Werte Test:", test.isnull().sum().sum())
```

```python
train['Sex'] = train['Sex'].map({'male': 0, 'female' : 1})
test['Sex'] = test['Sex'].map({'male': 0, 'female' : 1})


train = pd.get_dummies(train, columns=['Embarked'])
test  = pd.get_dummies(test, columns=['Embarked'])

print(train.head())
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score


X_train = train.drop(columns=['Survived'])
y_train = train['Survived']


model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

scores = cross_val_score(model, X_train, y_train, cv=5)
print("Genauigkeit:", scores.mean().round(3))
```

```python
# Vorhersagen auf Testdaten machen
X_test = test.drop(columns=['PassengerId'])
predictions = model.predict(X_test)


submission = pd.DataFrame({
    'PassengerId': test['PassengerId'],
    'Survived': predictions
})

submission.to_csv('submission.csv', index=False)
print(submission.head())
```
